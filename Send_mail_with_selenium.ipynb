```python
from selenium import webdriver  # driver selenium
import time  # manejo temporal
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By   # para elegir por elemento
from selenium.webdriver.support.ui import WebDriverWait # para el time sleep
from selenium.webdriver.support import expected_conditions as EC # condiciones esperadas
from selenium.webdriver import ActionChains as AC # acciones en cadena
# opciones del driver

opciones=Options() # charge options

opciones.add_argument('--start-maximized')  # empieza maximizado

opciones.headless=False # To see selenium actions 

opciones.add_experimental_option('excludeSwitches', ['enable-automation']) # para quitar flag de robot
#opciones.add_experimental_option('useAutomationExtension', False)

opciones.add_extension('../driver/adblock.crx')  # extension de adblocker

opciones.add_argument('--incognito') #access like unknow user

opciones.add_argument('user-data-dir=selenium')   # para guardar cookies


PATH=ChromeDriverManager().install()   #'driver/chromedriver'
```

```python
#lista de usuarios a los que queremos mandar el correo, hay que añadir una coma ',', al final de cada dirección.
mail_to = ['user1@user1.com,', 'user2@user2.com,','user2@user2.com,']
# mensaje que queremos envíar a los destinatarios
msg = 'Hello'
def send_mail(user, password, mail_lst, subject, msg):
    '''
        Funtion that send an email throught outlook app using selenium.
        
        Parameters.
        ---------------------------------------------------------------
        user = user email address to login in outlook. Type string.
        password = user password to login on outlook web. Type string.
        mail_lst = List of mail destinations. Type list.
        subject = Subject of the mail list. Type string.
        msg = Message that want to send to our mail list. Type string
        
        Returns
        ---------------------------------------------------------------
        
        It all are Ok, prints !Message or messages send correctly¡.
    
    '''
    url = 'https://outlook.live.com/owa/' # destination url
    driver=webdriver.Chrome(PATH, options=opciones) # charge selenium options
    driver.get(url) # go to url

    time.sleep(2)
    try: # if selenium finds cookies accept it
        cookies = driver.find_element_by_xpath('//*[@id="L2AGLb"]') # cookies box
        cookies.click() # accept cookies
    except:
        print('No cookies') # if no cookies print it

    mail = driver.find_element_by_xpath('/html/body/header/div/aside/div/nav/ul/li[2]/a')
    mail.click() # click on user box to iniciate login

    
    use_user = driver.find_element_by_xpath('//*[@id="i0116"]')
    use_user.send_keys(user) # send user name 

    time.sleep(3)

    next_bt = driver.find_element_by_xpath('//*[@id="idSIButton9"]')
    next_bt.click() # iniciate login

    time.sleep(2)
    
    passw = driver.find_element_by_xpath('//*[@id="i0118"]')
    passw.send_keys(password) # write password on password box

    time.sleep(2)
    next_bt2 = driver.find_element_by_xpath('//*[@id="idSIButton9"]')
    next_bt2.click() # next button click

    time.sleep(2)

    next_bt3 = driver.find_element_by_xpath('//*[@id="idBtn_Back"]')
    next_bt3.click() # choose no to close our session on outlook when close the web browser

    time.sleep(2)

    new_ms = driver.find_element_by_xpath('//*[@id="id__8"]')
    new_ms.click() # click on new message button

    time.sleep(2)
    
    select_cco = driver.find_element_by_xpath('//*[@id="ReadingPaneContainerId"]/div/div/div/div[1]/div[1]/div[1]/div/div[2]/button[2]/span/span')
    select_cco.click() # chosse CCO options to send our mails
    time.sleep(2)
    
    for mail in mail_lst: 
        #This loop iterate in the mail list and add a new mail direction on each iteration
        destination = driver.find_element_by_xpath('//*[@id="ReadingPaneContainerId"]/div/div/div/div[1]/div[1]/div[1]/div[2]/div/div/div/div/div/div[1]/div/div/input')
        destination.send_keys(mail)
        time.sleep(2)
    time.sleep(2)
    
    subject_box = driver.find_element_by_xpath('//*[@id="TextField524"]')
    subject_box.send_keys(subject) # write the subject of our mail list
    time.sleep(2)
    
    msg_box = driver.find_element_by_xpath('//*[@id="virtualEditScroller527"]/div[1]/div')
    msg_box.send_keys(msg) # write the message that want to send our destinators
    time.sleep(3)
    
    send_msg = driver.find_element_by_xpath('//*[@id="id__529"]')
    send_msg.click() # click on send button
    print('!Message or messages send correctly¡')
    time.sleep(5)
    driver.quit() # close driver
```
